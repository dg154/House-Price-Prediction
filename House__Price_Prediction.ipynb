```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

X=data.drop(['median_house_value'],axis=1)
y=data['median_house_value']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2)
```

```python
train_data= X_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.drop(['ocean_proximity'], axis=1, inplace=True)
```

```python
train_data.corr()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
train_data['total_rooms']= np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']= np.log(train_data['total_bedrooms']+1)
train_data['population']= np.log(train_data['population']+1)
train_data['households']= np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.insert(1, 'ocean_proximity', data['ocean_proximity'])
```

```python
train_data.ocean_proximity.value_counts()
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity, prefix="ocean_")).drop(['ocean_proximity'], axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=train_data, x="latitude", y="longitude", hue="median_house_value", palette="coolwarm" )
```

```python
train_data['bedroom_ration']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['households_rooms']=train_data['total_rooms']/train_data['households']
```

```python
from sklearn.linear_model import LinearRegression

X_train,y_train=train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']

reg=LinearRegression()
reg.fit(X_train, y_train)
```

```python
test_data= X_test.join(y_test)

test_data['total_rooms']= np.log(test_data['total_rooms']+1)
test_data['total_bedrooms']= np.log(test_data['total_bedrooms']+1)
test_data['population']= np.log(test_data['population']+1)
test_data['households']= np.log(test_data['households']+1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity, prefix="ocean_")).drop(['ocean_proximity'], axis=1)
test_data['bedroom_ration']=test_data['total_bedrooms']/test_data['total_rooms']
test_data['households_rooms']=test_data['total_rooms']/test_data['households']
```

```python
X_test,y_test=test_data.drop(['median_house_value'],axis=1),test_data['median_house_value']
```

```python
reg.score(X_test,y_test)
```
